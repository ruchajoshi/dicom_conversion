```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_nifti(file_path):
    img = nib.load(file_path)
    print(img)
    data = img.get_fdata()
    return data
```

```python
def compare_nifti(nifti1_path, nifti2_path):
    data1 = load_nifti(nifti1_path)
    data2 = load_nifti(nifti2_path)
    
    if np.array_equal(data1, data2):
        print("The NIfTI files are identical.")
    else:
        print("The NIfTI files are different.")
        diff = np.abs(data1 - data2)
        print(f"Max difference: {np.max(diff)}")
```

```python
def plot_slices(nifti1, nifti2, slice_idx=None, axis=2):
    """Plots a middle slice (or a given slice index) from two NIfTI images side by side."""
    if slice_idx is None:
        slice_idx = nifti1.shape[axis] // 3  # Default to middle slice
    
    slice1 = np.take(nifti1, slice_idx, axis=axis)
    slice2 = np.take(nifti2, slice_idx, axis=axis)
    
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(slice1.T, cmap='gray', origin='lower')
    axes[0].set_title("NIfTI 1")
    axes[0].axis("off")
    
    axes[1].imshow(slice2.T, cmap='gray', origin='lower')
    axes[1].set_title("NIfTI 2")
    axes[1].axis("off")
    
    plt.show()
```

```python
nifti_path1 = "labels_prostate/7139000004_V_prostate_mask.nii.gz"  # Replace with actual path
nifti_path2 = "prostate.nii"      # Replace with actual path
```

```python
data1 = load_nifti(nifti_path1)
data2 = load_nifti(nifti_path2)
plot_slices(data1, data2)
```

```python
compare_nifti(nifti_path1, nifti_path2)
```
